## Now we will try to check the normality of the numerical variables visually

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Use seaborn library to construct distribution plots for the numerical variables

```python
fig, axes = plt.subplots(4, 2,figsize=(15,15))
fig.suptitle('Numerical data for the dataframe')

for i,x in enumerate(numerical.columns):
    sns.distplot(numerical[x], ax=axes[int(i/2),i%2])
```

## Use Matplotlib to construct histograms

```python
fig, axs = plt.subplots(4, 2,tight_layout=True,figsize=(15,15))
for i,x in enumerate(numerical.columns):
    axs[int(i/2),i%2].hist(numerical[x])
```
